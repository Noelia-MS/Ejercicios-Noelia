### Ejercicio 26:
- Encontrar el año en que hubo más migración por continente y crear un dataframe con nombre de continente, el año de máxima migración y cantidad de migración.

|     |Continente|Año |Migraciones|
|-----|----------|----|-----------|
|**0**|Asia      |2010|163845     |
|**1**|Europe    |1990|68301      |
|**2**|Africa    |2010|40892      |
|**3**|Oceania   |1992|2871       |
|**4**|America   |1992|44699      |

```python
df_26 = df.replace({"Latin America and the Caribbean" : "America", 
                    "Northern America": "America"})
df_26["Continente"].unique()
```

```python
migraciones_por_continente = df_26.groupby('Continente').sum(numeric_only=True)
```

```python
resultados = []

for continente in migraciones_por_continente.index:
    max_migracion = migraciones_por_continente.loc[continente].max()
    año_max = migraciones_por_continente.loc[continente].idxmax()
    resultados.append([continente, año_max, max_migracion])
```

```python
df26_final = pd.DataFrame(resultados, columns=['Continente', 'Año', 'Migraciones'])
```

```python
df26_final
```
