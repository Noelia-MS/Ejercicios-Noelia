### Ejercicio 26:
- Encontrar el a√±o en que hubo m√°s migraci√≥n por continente y crear un dataframe con nombre de continente, el a√±o de m√°xima migraci√≥n y cantidad de migraci√≥n.

|     |Continente|A√±o |Migraciones|
|-----|----------|----|-----------|
|**0**|Asia      |2010|163845     |
|**1**|Europe    |1990|68301      |
|**2**|Africa    |2010|40892      |
|**3**|Oceania   |1992|2871       |
|**4**|America   |1992|44699      |

```python
df_26 = df.replace({"Latin America and the Caribbean" : "America", 
                    "Northern America": "America"})
df_26["Continente"].unique()
```

```python
migraciones_por_continente = df_26.groupby('Continente').sum(numeric_only=True)
```

```python
resultados = []

for continente in migraciones_por_continente.index:
    max_migracion = migraciones_por_continente.loc[continente].max()
    a√±o_max = migraciones_por_continente.loc[continente].idxmax()
    resultados.append([continente, a√±o_max, max_migracion])
```

```python
df26_final = pd.DataFrame(resultados, columns=['Continente', 'A√±o', 'Migraciones'])
```
