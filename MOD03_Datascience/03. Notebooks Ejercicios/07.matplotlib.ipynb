### Ejercicio 10:
- Muestra en 2 graficas de sectores con _**plt.subplots()**_ la siguiente información:
    1. En la primera gráfica mostrar el número de pasajeros que murieron de cada clase.
    2. En la segunda gráfica mostrar el número de pasajeros que sobrevivieron de cada clase.

<div align="center">
   
</div>

```python
labels= sorted([f"clase {i}" for i in df_titanic["Pclass"].unique()])

plt.subplot(1,2,1)
values= df_titanic[df_titanic["Survived"]== 0].value_counts("Pclass", sort=False, normalize=True)
plt.pie(x=values, labels=labels, autopct="%1.2f%%")


plt.subplot(1,2,2)
values= df_titanic[df_titanic["Survived"]== 1].value_counts("Pclass", sort=False, normalize=True);
plt.pie(x=values, labels=labels, autopct="%1.2f%%")
plt.title( "Not survived VS Survived %")
```

### Ejercicio 12:
- Muestra en una grafica de caja la edad de los que sobrevivieron y de los que no sobrevivieron.

**Nota**: Para poder graficarlo hay que reemplazar los NaN por el promedio de edad.

<div align="center">
   
</div>

```python
df_titanic["Age"]= df_titanic["Age"].fillna(df_titanic["Age"].median())
```

```python
# dos gráfico boxplot diferentes
plt.boxplot(x = df_titanic[df_titanic["Survived"]== 0]["Age"], tick_labels=["Not survived"]);
plt.show()
plt.boxplot(x = df_titanic[df_titanic["Survived"]== 1]["Age"], tick_labels=["Survived"]);
plt.show()
```

```python
# mismo bloxplot
etiquetas = ["Not survived", "Survived"]
data = [
    df_titanic[df_titanic["Survived"]== 0]["Age"],
    df_titanic[df_titanic["Survived"]== 1]["Age"],
]
plt.boxplot(x=data, tick_labels=etiquetas, showmeans=True);
plt.grid()
```

### Ejercicio 16:
- Muestra en un histograma la edad de los pasajeros que sobrevivieron y la edad de los que no.

<div align="center">
   
</div>

```python
plt.hist(df_titanic[df_titanic["Survived"]==0]['Age'], alpha=0.5, label='No supervivientes')
plt.hist(df_titanic[df_titanic["Survived"]==1]['Age'], alpha=0.5, label='Supervivientes')
plt.xlabel('Edad')
plt.ylabel('Pasajeros')
plt.legend()
plt.show()
```
