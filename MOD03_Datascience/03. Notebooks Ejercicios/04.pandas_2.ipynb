### Ejercicio 8:
- Remplaza los valores de sexo, por M para male y F para female.

```python
# Transformar datos 
# 1. Crear un diccionario de python con los valores a mapear, importante que coincidan con el valor existente o saldrá nan.
genre_dict = {
    "male": "M",
    "female": "F"}
# 2. Pasar el diccionario como parámetro de la función .map sobre la columna género
df["Sex"].map(genre_dict)
# Es un filtro, si queremos guardarlo asignamos una función
```

### Ejercicio 9:
- Convierte en mayusculas todos los nombres de los pasajeros.

```python
# apply: recibe una función def o una función lambda
def convertir_mayus(text):
    return text.upper()

    
df["Name"].apply(convertir_mayus)
```

```python
df["Name"].map(lambda name : name.upper())
```
